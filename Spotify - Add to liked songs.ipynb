```python
for pl_id in tuple(df_playlists.id):
    print(pl_id)
    limit = 50
    offset = 0
    num_tracks = 1
    while num_tracks > 0:
        playlist_items = (sp.playlist_items(pl_id, limit=limit, offset=offset) or {}).get("items", [])
        df_playlist_tracks = pd.DataFrame(playlist_items)
        playlist_tracks = df_playlist_tracks.track.apply(lambda e: e.get("id")).to_list()
        sp.current_user_saved_tracks_add(playlist_tracks)
        num_tracks = len(playlist_items)
        offset += limit
```
